```r
library(ggtree)

plot_tree <- function(gene_tree_file, plot_title, label_size = 2, right_margin = 3) {
    

  tree_path <- "/scratch/jpm73279/13.ACR_CNS_analysis/01.analysis/run_orthofinder/final.grasses_vs_athal/OrthoFinder/Results_Nov08/Orthogroups/c4_gene_trees/"
  gene_tree <- paste0(tree_path, gene_tree_file)
  # Read the tree file
  tree <- read.tree(gene_tree)
  
  # Set the size of the plot
  options(repr.plot.width=8, repr.plot.height=10)
  
  # Create the base tree plot with adjusted margin
  p <- ggtree(tree, branch.length="none") + 
    geom_tree() +
    theme_tree2() +  # Make sure to use the correct theme function available in your ggtree version
    geom_tiplab(size=label_size, hjust=0) +  # Adjust label size and horizontal justification
    theme(plot.margin=unit(c(1, right_margin, 1, 1), "cm")) +  # Adjust the margin
    coord_cartesian(clip="off") +  # Avoid clipping tip labels
    ggtitle(plot_title)  # Add the title
    
  # View the plot
  print(p)
}

# To use this function, call it with the file path and desired title like so:
# plot_tree("path/to/your/gene_tree_file.nwk", "Your Plot Title")
```
