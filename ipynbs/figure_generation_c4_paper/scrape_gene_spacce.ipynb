```r
library(tidyverse)
```

```r
sb_vs_zm <- read_delim("/scratch/jpm73279/16.GENESPACE/results/syntenicBlock_coordinates.csv", delim = ",", col_names = TRUE) %>% 
    separate(blkID, into = c("class", "syntenic_block_group"), sep = ": ") %>% 
    dplyr::filter(class == "Zm_vs_Sb")
```

```r
sb_vs_zm
```

```r
gene_key_file <- read_delim("/scratch/jpm73279/16.GENESPACE/results/SequenceIDs.txt", delim = ": ", col_names = c("ID", "gene_name"))
```

```r
head(gene_key_file)
```

```r
sb_vs_zm %>% 
    dplyr::filter(genome1 == "Sb") %>% 
    dplyr::select(chr1, startBp1, endBp1, firstGene1, lastGene1, syntenic_block_group, orient)
```

```r
sb_genome_1_grab <- sb_vs_zm %>% 
    dplyr::filter(genome1 == "Sb") %>% 
    dplyr::select(chr1, startBp1, endBp1, firstGene1, lastGene1, syntenic_block_group, orient) %>% 
    arrange(syntenic_block_group) %>% 
    dplyr::left_join(., gene_key_file, by = c("firstGene1" = "ID")) %>% 
    dplyr::rename("firstGeneBlock" = gene_name) %>% 
    dplyr::left_join(., gene_key_file, by = c("lastGene1" = "ID")) %>% 
    dplyr::rename("lastGeneBlock" = gene_name) %>% 
    dplyr::mutate(gene_names = str_c(firstGeneBlock, lastGeneBlock, sep = ";")) %>% 
    dplyr::select(chr1, startBp1, endBp1,gene_names,syntenic_block_group)
```

```r
sb_genome_1_grab
```

```r
zm_genome_2_grab <- sb_vs_zm %>% 
    dplyr::filter(genome2 == "Zm") %>% 
    dplyr::select(chr2, startBp2, endBp2, firstGene2, lastGene2, syntenic_block_group, orient) %>% 
  mutate(
    start = case_when(
      orient == "+" ~ startBp2,
      orient == "-" ~ endBp2
    ),
    end = case_when(
      orient == "+" ~ endBp2,
      orient == "-" ~ startBp2
    )
  ) %>% 
    arrange(syntenic_block_group) %>% 
    dplyr::left_join(., gene_key_file, by = c("firstGene2" = "ID")) %>% 
    dplyr::rename("firstGeneBlock" = gene_name) %>% 
    dplyr::left_join(., gene_key_file, by = c("lastGene2" = "ID")) %>% 
    dplyr::rename("lastGeneBlock" = gene_name) %>% 
    dplyr::mutate(gene_names = str_c(firstGeneBlock, lastGeneBlock, sep = ";")) %>% 
    dplyr::select(chr2, start, end,gene_names,syntenic_block_group)
```

```r
zm_genome_2_grab$start < zm_genome_2_grab$end
```

```r
write_tsv(zm_genome_2_grab, file = "/scratch/jpm73279/17.syntenic_block_exploration/zm.syntenic_blocks.bed", col_names = FALSE)
write_tsv(sb_genome_1_grab, file = "/scratch/jpm73279/17.syntenic_block_exploration/sb.syntenic_blocks.bed", col_names = FALSE)
```
